## 데이터셋 준비

```python
## 데이터셋 파일 경로 설정

path1 = "dataset/D1-Exp-and-Salary.txt"
path2 = "dataset/D1-Unknown-but-for-Regression.txt"
path3 = "dataset/D1-Pass-or-Fail.txt"
```

```python
with open(path1, "r") as f:
    lines = f.readlines()
    lines = list(map(lambda x: list(map(lambda y: float(y), x.strip().split(","))), lines))
```

```python
import pandas as pd

df = pd.read_csv(path2, sep=",", header=None)
df.head()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

데이터 전처리와 관련된 기본 연산

- 평균
- 표준편차
- 최소
- 최대

```python
import numpy as np

# axis=0: 세로로 합치기
X_mean = np.mean(X, axis=0)
y_mean = np.mean(y)

X_std = np.std(X, axis=0)
y_std = np.std(y)

X_min, X_max = np.min(X, axis=0), np.max(X, axis=0)
y_min, y_max = np.min(y), np.max(y)
```

정규화 1) min-max, normalization

```python
X_norm = (X - X_min)  / (X_max - X_min)
y_norm = (y - y_min) / (y_max - y_min)
X_norm, y_norm
```

정규화 2) 

```python
X_zscore = (X - X_mean)  / X_std
y_zscore = (y - y_mean)  / y_std

# 결과
print("X_mean:", X_mean)
print("X_std:", X_std)
print("X_min:", X_min)
print("X_max:", X_max)
print("X_norm:", X_norm)
print("X_zscore:", X_zscore)
```
